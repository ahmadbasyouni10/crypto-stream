```python
import requests
# website url with standings
standings_url = 'https://fbref.com/en/comps/12/2023-2024/2023-2024-La-Liga-Stats'
data = requests.get(standings_url)
data
```

```python
from bs4 import BeautifulSoup
import pandas as pd
# Use BeautifulSoup to parse HTML data of each team
soup = BeautifulSoup(data.text)
# css selector gives flexibility to select elements, id, class, etc.
# table.stats_table is the class of the table containing the standings
# [0] is used to select the first table with the class stats_table
standingsTable = soup.select('table.stats_table')[0]
standingsTable
df = pd.read_html(str(standingsTable))[0]
df
```

```python
# find all links in the table  (team links)
links = standingsTable.find_all('a')
# list comprenehsion to get the href attribute of each link in the list of links
links = [l.get('href') for l in links]
links = [l for l in links if '/squads' in l]
prefix = 'https://fbref.com'
# add prefix to each link to form the absolute url
links = [prefix +l for l in links]
links
```

```python
import pandas as pd 
real_madrid_url = links[0]
matches_RM_df = pd.read_html(real_madrid_url, attrs={'id':'matchlogs_for'})[0]
matches_RM_df
```

```python
data = requests.get(real_madrid_url)
soup = BeautifulSoup(data.text)
links = soup.find_all('a')
linksnew = [l.get('href') for l in links]
linksShooting = [l for l in linksnew if l and 'shooting' in l]
linksShooting
```

```python
data = requests.get(prefix + linksShooting[0])
shootingdf = pd.read_html(data.text, match="Shooting")[0]
shootingdf
```

```python
# drop multi-level columns to make it easier to work with using pandas
shootingdf.columns = shootingdf.columns.droplevel()
```

```python
# merge the two dataframes
# we only need the non overlapping columns from shooting df
columnsWanted = ['Date', 'Gls', 'Sh', 'Dist', 'FK', 'PK', 'PKatt']
RMdf = matches_RM_df.merge(shootingdf[columnsWanted], on='Date')
```
